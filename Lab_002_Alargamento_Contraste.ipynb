````python
# %% setup - Importações e funções utilitárias
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2 as cv
from pathlib import Path
````

````python
def converte_cinza(figura):
    """
    Recebe o caminho de uma imagem BMP, converte para escala de cinza, salva a imagem, retorna o caminho.
    """
    # Carrega a imagem BMP e converte para escala de cinza
    path = Path(figura)
    img = Image.open(figura).convert('L')

    # Salva a imagem em escala de cinza
    out_path = path.with_name(path.stem + "_cinza.bmp")
    img.save(out_path, format="BMP")

    return str(out_path)
````

## 2) Abrindo uma imagem e apresentando seu histograma
Vamos criar uma função **Python** que:
- **Receba** o caminho de uma imagem BMP em escala de cinza.
- **Apresente** a imagem e o histograma da imagem.
- **Use a imagem ``lena_lc.bmp``** disponibilizada no github (pasta ``data``) para os testes.

Nome sugerido para a função: `mostra_img_histograma(figura)`.

````python
def mostra_img_histograma(figura):
    """
    Recebe o caminho de uma imagem BMP,
    exibe a imagem e apresenta seu histograma (lado a lado).
    """

    # Carrega a imagem BMP
    path = Path(figura)
    img = Image.open(figura).convert('L')
    img_array = np.array(img).flatten()

    # Cria figura com 2 colunas
    plt.figure(figsize=(12,5))

    # Mostra a imagem
    plt.subplot(1,2,1)
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.title(f"Imagem (cinza): {path.name}")
    plt.axis('on')

    # Mostra o histograma
    plt.subplot(1,2,2)
    plt.hist(img_array, bins=256, range=(0,255), color='black')
    plt.title(f"Histograma ({path.name})")
    plt.xlabel("Intensidade (0–255)")
    plt.ylabel("Frequência de pixels")

    plt.tight_layout()
    plt.show()
````

# 🖼️ Tarefa A — Função de Alargamento de Contraste

Implemente uma função em **Python** que:  

1. **Receba** o caminho de uma imagem BMP (`figura`) em escala de cinza.  
2. **Calcule** os valores mínimo (`imin`) e máximo (`imax`) de intensidade presentes na imagem.  
3. **Reescale** (normalize) todos os pixels para que `imin` seja mapeado para `0` e `imax` seja mapeado para `255`.
4. **Salve** a nova imagem em um novo arquivo BMP.
5. **Retorne** o caminho da nova imagem.

⚠️ Caso `imin == imax` (imagem com intensidade constante), a função deve simplesmente retornar uma imagem preta (todos os valores 0).  

---

### Nome sugerido da função:
```python
def contraste(figura):
    ...


````python
import numpy as np
from PIL import Image
from pathlib import Path

def contraste(figura):
    """
    Recebe o caminho de uma imagem BMP já em escala de cinza,
    aplica alargamento de contraste (min -> 0, max -> 255),
    salva a nova imagem em formato BMP
    e retorna o caminho do arquivo salvo.
    """

    # 1. Carrega a imagem
    img = Image.open(figura).convert('L')
    img_np = np.array(img).astype(np.float32)

    # 2. Calcula valores mínimo e máximo
    imin = np.min(img_np)
    imax = np.max(img_np)

    # 3. Alargamento de contraste
    if imin == imax:
        img_out = np.zeros_like(img_np, dtype=np.uint8)
    else:
        img_out = (img_np - imin) * (255.0 / (imax - imin))
        img_out = img_out.astype(np.uint8)

    # 4. Salvar nova imagem BMP
    path = Path(figura)
    novo_caminho = path.parent / f"{path.stem}_contraste.bmp"
    Image.fromarray(img_out).save(novo_caminho)
    return str(novo_caminho)
````

````python
# Imagem original
img_original = '/content/lena_lc.bmp'

# Mostra original
mostra_img_histograma(img_original)

# Aplica contraste
img_contraste = contraste(img_original)

# Mostra imagem com contraste
mostra_img_histograma(img_contraste)
````

## 📝 Tarefa C — Realce de Contraste com Transformação de Potência

Nesta atividade, você irá aplicar uma técnica de Transformação de Potência para o realce de contraste em uma imagem, implementada e disponível nas bibliotecas OpenCV e NumPy (cv2 e numpy, importadas neste Notebook).

A transformação de potência (também conhecida como correção gama) é uma técnica de processamento de imagem usada para ajustar o contraste, realçando detalhes em áreas escuras ou claras.

Fórmula: **s = c$r^{γ}$**

Onde:

r: é o pixel de entrada;

s: a saída;

c: uma constante positiva (geralmente 1); e

γ (gama): é o parâmetro de ajuste.

Sendo:

* **γ < 1 (0.1 - 0.9)**: mapeiam poucos valores escuros para muitos valores na saída (aumenta o brilho de áreas escuras).

* **γ > 1 (1.1 - 5.0)**: mapeiam muitos valores escuros para poucos valores na saída (reduz o brilho de áreas claras).

* **γ = 1**: temos a transformação de identidade (sem efeito).

Para esta tarefa, lembre-se:

1. **img / 255.0**: NumPy converte a imagem para float e normaliza. Operações de potência em inteiros (uint8) podem resultar em erros.

2. **np.power(array, gama)**: Função do NumPy que aplica a potência a cada elemento da matriz (pixel) de forma otimizada.

3. **.astype(np.uint8)**: Reconverte os valores para 0-255 inteiros, necessários para o OpenCV exibir a imagem.

EXERCÍCIO: **Aplique a técnica de transformação de potência para realçar o contraste de uma imagem em BMP (lena_lc.bmp), em escala de cinza, e altere o valor em γ (gama) para visualizar os efeitos sobre a imagem.**

````python
# Carregar a imagem em escala de cinza
img_cinza = converte_cinza('lena_lc.bmp')

# se a função retornar o nome do arquivo em vez da imagem, carregamos a imagem corretamente em escala de cinza
if isinstance(img_cinza, str):
    img_cinza = cv.imread(img_cinza, cv.IMREAD_GRAYSCALE)

# Normalizar os valores dos pixels para o intervalo [0, 1]
img_norm = img_cinza.astype(np.float32) / 255.0

# Aplicar a transformação de potência (gama)
gamma = 0.9
img_gamma = np.power(img_norm, gamma)

# Converter de volta para o intervalo [0, 255] e tipo uint8
img_saida = (img_gamma * 255).astype(np.uint8)

# Exibir os resultados ao final
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(img_cinza, cmap='gray', vmin=0, vmax=255)
plt.title("Original (cinza)")
plt.axis('on')

plt.subplot(1,2,2)
plt.imshow(img_saida, cmap='gray', vmin=0, vmax=255)
plt.title(f"Gama = {gamma}")
plt.axis('on')

plt.tight_layout()
plt.show()
````
